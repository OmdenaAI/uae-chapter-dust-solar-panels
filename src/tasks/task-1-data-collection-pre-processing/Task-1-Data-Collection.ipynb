```python
def Convert_To_RGB(folder):
    for image in (os.listdir(folder)):
        img = Image.open(folder+ "/" + image)
        if str(img.mode) != 'RGB':
            rgba_image = Image.open(folder+ "/" + image)
            rgb_image = rgba_image.convert('RGB')
            old_name = os.path.join(folder, image)
            only_name = os.path.splitext(image)[0]
            new_base = only_name + '.jpg'
            new_name = os.path.join(folder, new_base)
            rgb_image.save(new_name, format='JPEG', quality=100)
            os.remove(old_name)

Convert_To_RGB(Path)
```

## Filter images by size.

```python
def filterImages(path, thresholdWidth, thresholdHeight):
    imgs = []
    valid_images = [".jpg", ".jpeg", ".JPG", ".JPEG"]

    for f in os.listdir(path):
        ext = os.path.splitext(f)[1]

        if ext.lower() not in valid_images:
            continue
        imgs.append(f)

    directory = os.path.dirname(path + "filteredImages")
    if not os.path.exists(directory):
        os.makedirs(directory)

    filteredImages = []
    imgsToRemove = []

    for i in imgs:
        image = Image.open(os.path.join(path, i))
        width, height = image.size
        if thresholdHeight > height or thresholdWidth > width:
            imgsToRemove.append(i)

    for image in imgsToRemove:
        os.rename(os.path.join(path + "/" + image),
                  os.path.join(path + '/filteredImages/' + image))
        filteredImages.append(i)
    return filteredImages
```
